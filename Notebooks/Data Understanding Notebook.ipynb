```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = 'Customer_Dataset.xlsx'
data_df = pd.read_excel(file)
```

```python
data_df_x = data_df.drop(['CROSS_SELL_SUCCESS', 'NAME', 'EMAIL', 'FIRST_NAME', 'FAMILY_NAME'], axis = 1)
data_df_y = data_df.loc[ : , 'CROSS_SELL_SUCCESS']
```

```python
cur_column = 'REVENUE'
lower_bound = np.percentile(data_df_x[cur_column], 1)
upper_bound = np.percentile(data_df_x[cur_column], 99)

outliers = data_df_x[(data_df_x[cur_column] < lower_bound) | (data_df_x[cur_column] > upper_bound)]
normal = data_df_x[(data_df_x[cur_column] >= lower_bound) & (data_df_x[cur_column] <= upper_bound)]

sns.set(font_scale=1)
plt.figure(figsize=(10, 6))
plt.title("Distribution of revenue", fontsize=20)
plt.xlabel("Revenue", fontsize=10)
plt.ylabel("Count", fontsize=10)

sns.histplot(normal[cur_column], bins = 30, color = 'cyan', edgecolor='black', alpha = 0.6, label = 'Normal')
sns.scatterplot(x=outliers[cur_column], y = np.zeros_like(outliers[cur_column]), color = 'red', label='Outliers', zorder=5)

plt.axvline(x=lower_bound, color='red', linestyle='--', label='Lower Bound')
plt.axvline(x=upper_bound, color='red', linestyle='--', label='Upper Bound')
plt.legend(loc='upper right')
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()
```

```python

# Vẽ bản đồ nhiệt (heatmap) của ma trận tương quan (correlation matrix) giữa các trường dữ liệu.
numeric_cols = data_df.select_dtypes(include=['number'])
sns.set(font_scale=0.8)
corr_matrix = numeric_cols.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Heat map', fontsize = 20)
plt.show()

```

```python
int_cols = data_df.loc[ : , ['MOBILE_LOGINS', 'PC_LOGINS', 'CONTACTS_W_CUSTOMER_SERVICE', 'AVG_CLICKS_PER_VISIT']]
for col in int_cols.columns:
    plt.figure(figsize=(10, 6))
    sns.set(font_scale=1)
    sns.countplot(x=col, data=data_df_x, palette = 'viridis')
    plt.title(col, fontsize = 20)
    plt.xlabel(col, fontsize=10)
    plt.ylabel('Amount', fontsize=10)
    plt.show()
```

```python
int_cols = data_df.loc[ : , ['CANCELLATIONS_BEFORE_NOON', 'LATE_DELIVERIES']]
for col in int_cols.columns:
    sns.set(font_scale=1)
    plt.figure(figsize=(20, 8))
    sns.boxplot(x=col, y='CROSS_SELL_SUCCESS', data=data_df, palette = 'Set2')
    plt.title(col, fontsize=30)
    plt.xlabel(col, fontsize=20)
    plt.ylabel('CROSS_SELL_SUCCESS', fontsize=20)
    plt.show()
```

```python
int_cols = data_df.loc[ : , ['MOBILE_NUMBER',
                             'TASTES_AND_PREFERENCES']]
for col in int_cols.columns:
    plt.figure(figsize=(10, 6))
    cols = data_df.loc[ : , [col, 'CROSS_SELL_SUCCESS']]
    cross_tab = pd.crosstab(cols[col], cols['CROSS_SELL_SUCCESS'])

    count_df = cols.groupby([col, 'CROSS_SELL_SUCCESS']).size().reset_index(name='count')

    # Vẽ biểu đồ cột nhóm
    sns.barplot(data=count_df, x=col, y='count', hue='CROSS_SELL_SUCCESS', palette='pastel')

    plt.xlabel(col)
    plt.ylabel("Amount")
    plt.title("Distribution of "+col+" and CROSS_SELL_SUCCESS")
    plt.legend(title="CROSS_SELL_SUCCESS")
    plt.show()
```
